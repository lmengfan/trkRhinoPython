```python
import numpy as np
import pandas as pd
import json
import qgrid
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
f_in = open(r"D:\GitHub\trkRhinoPython\datastore.json", 'r')
datastore = json.load(f_in)
```

```python
# df = pd.DataFrame.from_records(objsdict)
df = pd.DataFrame(datastore)

df = df.convert_dtypes()
df[['areapy', 'area', 'level']] = df[['areapy', 'area', 'level']].apply(pd.to_numeric)
```

```python
df['levelname'] = ['지상{}층'.format(x) if y == 'ag' else '지하{}층'.format(abs(x)) for x,y in zip(df['level'], df['grade'])]
```

```python
pt = pd.pivot_table(df, values=['area', 'areapy'], index=['usage'], columns=['class'], aggfunc=np.sum)
pt.columns = pt.columns.swaplevel(0,1)
pt = pt.sort_index(axis=1)
pt
```

```python
unitcount = pd.pivot_table(df, values='uid', index=['usage'], columns=['class'], aggfunc='nunique')
unitcount
```

```python
areabylevel = pd.pivot_table(df, values=['area', 'areapy'], index=['level'], columns=['usage', 'class'], aggfunc=np.sum)
areabylevel = areabylevel.swaplevel(1,0, axis=1).sort_index(axis=1)
areabylevel
```
